## Data Cleaning

```python
# These are to be worked on:

# drop row id
# convert 'order-date' and 'ship-date' to datetime format
# customer id remains an object but special character '-' is removed
# orderId special character '-' is removed
# productId special characer '-' is removed
# fill up 11 missing entries in Postal Code column
# change 'Postal code' to string
```

```python
# save a copy of the data
df_copy = df.copy()
```

```python
# drop row id
df.drop(columns=['Row ID'], inplace=True)
```

```python
df.head(3)
```

```python
# fill up empty postal code entries
np.random.seed(42) # Set random seed for reproducibility

# Get the list of existing postal codes (drop missing ones)
postal_code_choices = df['Postal Code'].dropna().unique()

# Fill missing postal codes with a random choice from the list
df['Postal Code'] = df['Postal Code'].apply(
    lambda x: np.random.choice(postal_code_choices) if pd.isna(x) else x
)
```

```python
df.isnull().sum()
```

```python
# Converting Postal Code to string to preserve consistency (e.g., leading zeros)
df['Postal Code'] = df['Postal Code'].astype(str)
```

```python
df['Order ID'].dtypes
```

```python
df.dtypes
```

```python
# date conversion for 'orderDate' and 'shipDate'

# Converting Order Date and Ship Date to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d/%m/%Y')
```

```python
# duplicates showed a result of '1'
duplicates = df.duplicated()
duplicates.sum()
```

```python
# Remove exact duplicate rows
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
# Strip leading/trailing spaces and standardize text case for object columns
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.strip()      # remove extra spaces
    # df[col] = df[col].str.title()      # convert to title case (e.g., 'california' → 'California')
    df[col] = df[col].str.lower()        # convert to lower case
```

```python
# 1) Converting Order Date to datetime

df['Order Date'] = pd.to_datetime(df['Order Date'])


# 2) Extracting useful date parts
df['Order_Year']       = df['Order Date'].dt.year
df['Order_Month']      = df['Order Date'].dt.month                      # 1..12
df['Order_MonthName']  = df['Order Date'].dt.month_name()               # 'January', 'February',...
df['Order_DayOfWeek']  = df['Order Date'].dt.day_name()                 # 'Monday', 'Tuesday',...
df['Order_MonthPeriod'] = df['Order Date'].dt.to_period('M')            # pandas Period('2020-01', 'M')


# 3) Monthly aggregations (sales & order count)
monthly = (
    df.groupby('Order_MonthPeriod')
      .agg(Total_Sales=('Sales', 'sum'),
           Order_Count=('Order ID', 'count'))   
      .reset_index()
      .sort_values('Order_MonthPeriod')
)


# For a string column for plotting/labels:
monthly['Month_Year'] = monthly['Order_MonthPeriod'].astype(str)  # e.g., '2020-01'
```

```python
# What this (above) gives and why it matters:
# 'Order_Month' Period is perfect for chronological grouping and guaranteed to sort correctly.
# 'monthly' now contains total sales per month and the number of order rows per month; good for trend lines and seasonal checks.
# 'Order_Year', 'Order_MonthName', and 'Order_DayOfWeek' allows year-over-year, month-of-year, or weekday analyses easily.
```

## Visualisations

### Time Analysis

```python
# Simple Total Sales line plot for the monthly dataframe

plt.figure(figsize=(12, 6))
plt.plot(monthly['Month_Year'], monthly['Total_Sales'], marker='o', color='teal')
plt.title('Total Sales Over Time', fontsize=16)
plt.xlabel('Month-Year', fontsize=12)
plt.ylabel('Total Sales ($)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
# save as image ......................................
# plt.savefig('images/monthly_sales.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Order Count as a second line on the same plot so as to compare sales vs. orders

plt.figure(figsize=(12, 6))
plt.plot(monthly['Month_Year'], monthly['Total_Sales'], marker='o', color='teal', label='Total Sales ($)')
plt.plot(monthly['Month_Year'], monthly['Order_Count'], marker='s', color='orange', label='Order Count')

plt.title('Total Sales & Order Count Over Time', fontsize=16)
plt.xlabel('Month-Year', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.xticks(rotation=45)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# This (above) lets us see:
# If spikes in sales were driven by more orders - this is the case above
# Or by higher order values in certain months
```

```python
# Highlighting spikes and dips so the chart becomes more storytelling-friendly

plt.figure(figsize=(12, 6))
plt.plot(monthly['Month_Year'], monthly['Total_Sales'], marker='o', color='teal', label='Total Sales ($)')
plt.plot(monthly['Month_Year'], monthly['Order_Count'], marker='s', color='orange', label='Order Count')


# Highlight max & min sales points
max_sales_idx = monthly['Total_Sales'].idxmax()
min_sales_idx = monthly['Total_Sales'].idxmin()

plt.scatter(monthly.loc[max_sales_idx, 'Month_Year'], monthly.loc[max_sales_idx, 'Total_Sales'],
            color='green', s=120, zorder=5, label='Peak Sales')
plt.scatter(monthly.loc[min_sales_idx, 'Month_Year'], monthly.loc[min_sales_idx, 'Total_Sales'],
            color='red', s=120, zorder=5, label='Lowest Sales')


# Annotate
plt.annotate(f"Peak: ${monthly.loc[max_sales_idx, 'Total_Sales']:.0f}",
             xy=(monthly.loc[max_sales_idx, 'Month_Year'], monthly.loc[max_sales_idx, 'Total_Sales']),
             xytext=(0, 20), textcoords='offset points', ha='center', fontsize=10,
             arrowprops=dict(facecolor='green', arrowstyle='->'))

plt.annotate(f"Low: ${monthly.loc[min_sales_idx, 'Total_Sales']:.0f}",
             xy=(monthly.loc[min_sales_idx, 'Month_Year'], monthly.loc[min_sales_idx, 'Total_Sales']),
             xytext=(0, -30), textcoords='offset points', ha='center', fontsize=10,
             arrowprops=dict(facecolor='red', arrowstyle='->'))


# plot
plt.title('Total Sales & Order Count Over Time (with Highlights)', fontsize=16)
plt.xlabel('Month-Year', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.xticks(rotation=45)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# This (above):
# Puts green and red dots on the highest and lowest sales months
# Adds arrows & labels for ease of noticing key points
```

```python
# Seasonal Patterns: finding which months tend to perform best/worst overall, regardless of year

# Extracting month number & name
monthly['Month'] = pd.to_datetime(monthly['Month_Year']).dt.month
monthly['Month_Name'] = pd.to_datetime(monthly['Month_Year']).dt.strftime('%b')


# Grouping across years
seasonal = monthly.groupby(['Month', 'Month_Name']).agg({
    'Total_Sales': 'mean',
    'Order_Count': 'mean'
}).reset_index().sort_values('Month')


# Plot seasonal pattern
plt.figure(figsize=(10, 5))
plt.plot(seasonal['Month_Name'], seasonal['Total_Sales'], marker='o', color='teal', label='Avg Sales ($)')
plt.plot(seasonal['Month_Name'], seasonal['Order_Count'], marker='s', color='orange', label='Avg Orders')

plt.title('Average Monthly Sales & Orders Across All Years', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Average Values', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()

seasonal
```

### Category & Segment (Product) analysis

-  to see what’s driving sales; uncovering where the money is coming from.

```python
# Sales & Order Counts by Category

category_summary = df.groupby('Category').agg({
    'Sales': 'sum',
    'Order ID': 'nunique'
}).reset_index()

category_summary.rename(columns={'Order ID': 'Order_Count'}, inplace=True)

# Sort by sales
category_summary = category_summary.sort_values('Sales', ascending=False)

# Plot
plt.figure(figsize=(8,5))
sns.barplot(data=category_summary, x='Category', y='Sales', palette='viridis', hue='Category', legend=False)
plt.title('Total Sales by Category', fontsize=16)
plt.ylabel('Total Sales ($)')
plt.xlabel('Category')
plt.tight_layout()
plt.show()

category_summary
```

```python
# Segment Analysis (Consumer, Corporate, Home Office) to see who is/what group is buying the most; driving sales.

segment_summary = df.groupby('Segment').agg({
    'Sales': 'sum',
    'Order ID': 'nunique'
}).reset_index()

segment_summary.rename(columns={'Order ID': 'Order_Count'}, inplace=True)
segment_summary = segment_summary.sort_values('Sales', ascending=False)

# Plot
plt.figure(figsize=(8,5))
sns.barplot(data=segment_summary, x='Segment', y='Sales', hue='Segment', palette='coolwarm', legend=False)
plt.title('Total Sales by Segment', fontsize=16)
plt.ylabel('Total Sales ($)')
plt.xlabel('Segment')
plt.tight_layout()
plt.show()

segment_summary
```

```python
# Category & Segment into a single visualization (stacked bar) so we can see which customer type buys what category most.

cat_seg_summary = df.groupby(['Category', 'Segment']).agg({
    'Sales': 'sum'
}).reset_index()

# Pivot for stacked bar
cat_seg_pivot = cat_seg_summary.pivot(index='Category', columns='Segment', values='Sales')

# Plot
cat_seg_pivot.plot(kind='bar', stacked=True, figsize=(8,5), colormap='coolwarm')
plt.title('Sales by Category and Segment', fontsize=16)
plt.ylabel('Total Sales ($)')
plt.xlabel('Category')
plt.xticks(rotation=0)
plt.legend(title='Segment')
plt.tight_layout()
# save as image ...................................
# plt.savefig('images/category_sales.png', dpi=300, bbox_inches='tight')
plt.show()

cat_seg_summary
```

```python
# This (above) will help answer:
# Which segment dominates each category?
# Are there categories where sales are evenly distributed across segments?
```

### Shipping Analysis

```python
# Orders by Ship Mode: this will show which shipping methods are most frequently used

ship_mode_counts = df['Ship Mode'].value_counts().reset_index()
ship_mode_counts.columns = ['Ship Mode', 'Order Count']
ship_mode_counts
```

```python
# plot

plt.figure(figsize=(6, 6))
plt.pie(
    ship_mode_counts['Order Count'],
    labels=ship_mode_counts['Ship Mode'],
    autopct='%1.1f%%',
    startangle=90,
    colors=sns.color_palette('Set2')
)
plt.title('Distribution of Orders by Ship Mode', fontsize=14)
plt.show()
```

```python
# Average Shipping Time

df['Shipping_Days'] = (df['Ship Date'] - df['Order Date']).dt.days
avg_shipping_time = df['Shipping_Days'].mean()
avg_shipping_time
```

```python
# plot

plt.figure(figsize=(8, 5))
sns.histplot(df['Shipping_Days'], bins=10, kde=True, color='skyblue')
plt.title('Distribution of Shipping Times (Days)', fontsize=14)
plt.xlabel('Shipping Days')
plt.ylabel('Number of Orders')
plt.show()
```

```python
# Shipping Time by Ship Mode
# This step helps us see if different shipping methods actually deliver faster (or slower).

shipping_by_mode = df.groupby('Ship Mode')['Shipping_Days'].mean().reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(data=shipping_by_mode, x='Ship Mode', y='Shipping_Days', hue='Ship Mode', palette='viridis', legend=False)
plt.title('Average Shipping Days by Ship Mode', fontsize=14)
plt.xlabel('Ship Mode')
plt.ylabel('Average Shipping Days')
plt.show()
```
